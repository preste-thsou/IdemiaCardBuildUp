```python
import numpy as np
import matplotlib.pyplot as plt 
from problem.utils import GetDataFromFileFUN
```

```python
def plot_fronts(data_list, algos):
    
    def UpadeMax(data, max_x_T, max_y_C):

        if np.max(data[:,0]) > max_x_T:
            max_x_T = np.max(data[:,0])

        if np.max(data[:,1]) > max_y_C:
            max_y_C = np.max(data[:,1])

        return max_x_T, max_y_C


    fig = plt.figure(figsize = (15,7))
    ax = fig.add_subplot(1, 1, 1)
    
    max_x_T = 0
    max_y_C = 0

    for data_t, algo in zip(data_list, algos):
        data = np.array(data_t)
        data_T = data[:,0]
        data_C = data[:,1]

        ax.scatter(data_T, data_C,  s = 200, alpha = 0.7, label = "{}".format(algo))

        max_x_T, max_y_C = UpadeMax(data, max_x_T, max_y_C)



    ax.grid(True)
    ax.set_ylabel('OverallCost', fontsize=20)
    ax.set_xlabel('Tardiness', fontsize=20)
    ax.legend(fontsize = 15)
   

    st_x_T = int( (max_x_T + 5) / 10 )
    st_y_C = int( (max_y_C + 5) / 10 )
    

    ax.xaxis.set_ticks(np.arange(int(max_x_T*2.5/4), max_x_T + st_x_T + 5, max(1,st_x_T)))
    ax.yaxis.set_ticks(np.arange(int(max_y_C*2.5/4), max_y_C + st_y_C + 5, max(1,st_y_C)))

    plt.show()
    #plt.savefig(folder+'fronts_comparison.png')
```
