```python
import numpy as np
import matplotlib.pyplot as plt 
import copy
from problem.utils import BuildTotalPlanWithMHd, BuildTotalPlanWithMHs, AreAllOperationsRealized, AreAllOperationCompleted
from problem import problem_variables
from visualization import GetListOfSolutionsVar, PlotTimeplan, PlotTimeplanMH
```

# Fill a file name with solutions for visualization 

```python
vars_list = GetListOfSolutionsVar('/home/nazkam/Documents/Project_UTC/result/test_MHv5d_100k_corrected/SMPSO_1/VAR_front.SMPSO.mainproblem')
```

# Choose a solution to visualize

```python
nb_solution = len(vars_list)

nb_sol_to_viz = -1
while True:
    nb_sol_to_viz = input("What solution do you want to visualize ? (max:{}) \n- ".format(nb_solution))
    try:
        nb_sol_to_viz = int(nb_sol_to_viz) - 1
        break
    except:
        continue
```

# Choose a strategy  

```python
strategy = -1
while True:
    strategy = input("What strategy ? [1/2]\n- ")
    try:
        strategy = int(strategy)
        break
    except:
        continue
```

# Visualize

```python
solution = vars_list[nb_sol_to_viz]

Plan = copy.deepcopy(problem_variables.PLAN)
Settings = copy.deepcopy(problem_variables.SETTINGS)

if strategy == 1:
    BuildTotalPlanWithMHd(solution, Plan, Settings)
elif strategy == 2:
    BuildTotalPlanWithMHs(solution, Plan, Settings)

print('Are all operations realized ?   - ', AreAllOperationsRealized(Plan, Settings))
print('Are all operation completed ?   - ', AreAllOperationCompleted(Plan, Settings))

PlotTimeplanMH(Plan, Settings)
```
